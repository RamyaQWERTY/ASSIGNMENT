```python
from bs4 import BeautifulSoup
import requests
import re
from tabulate import tabulate


class CountryInformationScraper:
    def __init__(self, url):
        self.url = url
        self.response = None
        self.soup = None
        self.href_values = []

    def scrape_country_information(self):
        self.send_request()
        self.parse_html()
        self.extract_href_values()
        data = self.extract_country_information()
        self.display_country_information(data)

    def send_request(self):
        self.response = requests.get(self.url)

    def parse_html(self):
        self.soup = BeautifulSoup(self.response.content, 'html.parser')

    def extract_href_values(self):
        td_tags = self.soup.find_all("td", style="vertical-align:top;")
        for td_tag in td_tags:
            a_tag = td_tag.find("a")
            if a_tag:
                self.href_values.append(a_tag["href"])

    def extract_country_information(self):
        data = []

        for link in self.href_values:
            new_webpage = requests.get("https://en.wikipedia.org" + link)
            new_soup = BeautifulSoup(new_webpage.content, "html.parser")

            country_title = self.get_title(new_soup)
            country_info = self.get_information(new_soup)

            data.append([country_title, country_info])

        return data

    @staticmethod
    def get_title(soup):
        try:
            title = soup.find("span", attrs={'class': 'mw-page-title-main'}).text
        except AttributeError:
            title = ""
        return title

    @staticmethod
    def get_information(soup):
        try:
            a = soup.find('div', attrs={'id': 'mw-content-text'}).findChild('div', attrs={'class': 'mw-parser-output'}).find_all('p')
            text_list = ""
            for p_tag in a:
                text = p_tag.get_text(strip=True)
                if text:
                    text = re.sub(r'\[[^\]]*\]', '', text)  # Remove text within square brackets
                    text = re.sub(r'\([^)]*\)', '', text)  # Remove text within parentheses
                    text = re.sub(r'[\u2013\u2014]', '-', text)  # Replace em dashes and en dashes with hyphens
                    text = re.sub(r'\s+', ' ', text)
                    text = re.sub(r'^\W+|\W+$', '', text)
                    text_list += text
        except AttributeError:
            text_list = ""
        return text_list[:500]

    @staticmethod
    def display_country_information(data):
        headers = ["Country", "Information"]
        print(tabulate(data, headers, tablefmt="grid"))


# Create an instance of the CountryInformationScraper class
scraper = CountryInformationScraper("https://en.wikipedia.org/wiki/List_of_sovereign_states")

# Call the method to scrape and display country information
scraper.scrape_country_information()
```
