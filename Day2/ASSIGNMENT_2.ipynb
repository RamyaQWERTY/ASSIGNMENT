3.Extracted the introduction part (first paragraph) of each country's Wikipedia page and save the
collected information in a CSV file in the given format

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

URL = "https://en.wikipedia.org/wiki/List_of_sovereign_states"

# Function to extract Title
def get_title(soup):
    try:
        title = soup.find("span", attrs={'class': 'mw-page-title-main'}).text
    except AttributeError:
        title = ""
    return title

# Function to extract Information
def get_information(soup):
    try:
        a = soup.find('div', attrs={'id': 'mw-content-text'}).findChild('div', attrs={'class': 'mw-parser-output'}).find_all('p')
        text_list = ""
        for p_tag in a:
            text = p_tag.get_text(strip=True)
            if text:
                text = re.sub(r'\[[^\]]*\]', '', text)  # Remove text within square brackets
                text = re.sub(r'\([^)]*\)', '', text)  # Remove text within parentheses
                text = re.sub(r'[\u2013\u2014]', '-', text)  # Replace em dashes and en dashes with hyphens
                text = re.sub(r'\s+', ' ', text)
                text = re.sub(r'^\W+|\W+$', '', text)
                text_list += text
    except AttributeError:
        text_list = ""
    return text_list[:500]


if __name__ == '__main__':

    # add your user agent

    response = requests.get(URL)
    soup = BeautifulSoup(response.content, 'html.parser')
    td_tags = soup.find_all("td", style="vertical-align:top;")

    href_values = []
    for td_tag in td_tags:
        a_tag = td_tag.find("a")
        if a_tag:
            href_values.append(a_tag["href"])

    d = {"Country": [], "Information": []}

    # Loop for extracting product details from each link
    for link in href_values:
        new_webpage = requests.get("https://en.wikipedia.org" + link)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['Country'].append(get_title(new_soup))
        d['Information'].append(get_information(new_soup))

    new_df = pd.DataFrame.from_dict(d)
    new_df.to_csv("new.csv", header=True, index=False, sep=",")
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import sys

URL = "https://en.wikipedia.org/wiki/List_of_sovereign_states"

# Function to extract Title
def get_title(soup):
    try:
        title = soup.find("span", attrs={'class': 'mw-page-title-main'}).text
    except AttributeError:
        title = ""
    return title

# Function to extract Information
def get_information(soup):
    try:
        infobox = soup.find("table", class_="infobox")
        rows = infobox.find_all("tr")
        info_dict = {}
        for row in rows:
            header = row.find("th")
            if header:
                key = header.text.strip()
                value = row.find("td")
                if value:
                    info_dict[key] = value.text.strip()
        info_text = "\n".join(f"{key}: {value}" for key, value in info_dict.items())
        return info_text
    except AttributeError:
        return ""

if __name__ == '__main__':
    # add your user agent

    response = requests.get(URL)
    soup = BeautifulSoup(response.content, 'html.parser')
    td_tags = soup.find_all("td", style="vertical-align:top;")

    href_values = []
    for td_tag in td_tags:
        a_tag = td_tag.find("a")
        if a_tag:
            href_values.append(a_tag["href"])

    d = {"Country": [], "Information": []}

    # Loop for extracting product details from each link
    for link in href_values:
        new_webpage = requests.get("https://en.wikipedia.org" + link)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['Country'].append(get_title(new_soup))
        d['Information'].append(get_information(new_soup))

    df = pd.DataFrame.from_dict(d)
    df.to_csv(sys.stdout, index=False)
```
